##Performance vs Popularity

```python
cuisine_perf = df.groupby('Cuisine').agg({
    'Dining_Rating': 'mean',
    'Delivery_Rating': 'mean',
    'Average_Rating': 'mean',
    'Dining_Votes': 'sum',
    'Delivery_Votes': 'sum',
    'Total_Votes': 'sum'
}).reset_index()
cuisine_perf
```

```python
cuisine_perf['Popularity_Score'] = (
    cuisine_perf['Dining_Votes'] + cuisine_perf['Delivery_Votes']
)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=cuisine_perf,
    x='Average_Rating',
    y='Total_Votes',
    size='Popularity_Score',
    hue='Delivery_Rating',
    alpha=0.8,
    palette='viridis',
    legend=False
)
plt.title('Cuisine Performance vs Popularity')
plt.xlabel('Average Rating (Quality)')
plt.ylabel('Total Votes (Demand)')
plt.show()
```

```python
top10 = cuisine_perf.sort_values('Popularity_Score', ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(data=top10, x='Cuisine', y='Average_Rating', hue='Delivery_Rating')
plt.title('Top 10 Cuisines by Popularity and Average Rating')
plt.xticks(rotation=45)
plt.show()
```

Insights for Chef Innovator:

High Rating + High Demand (Safe & Profitable):

 North Indian,  Desserts, Beverages,  Chinese

Moderate Demand + High Rating (Emerging Opportunities):

 Sichuan,  South Indian

High Demand + Moderate Rating (Mass Market Improvement Areas):

 Pizza,  Fast Food, Biryani

##Market Saturation

```python
cuisine_sat = df.groupby('Cuisine').agg({
    'Restaurant_Name': 'nunique',
    'Avg_Rating_Cuisine': 'mean',
    'Avg_Price_Cuisine': 'mean',
    'Total_Votes': 'sum'
}).reset_index()

cuisine_sat.rename(columns={
    'Restaurant_Name': 'Restaurant_Count'
}, inplace=True)
cuisine_sat
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=cuisine_sat,
    x='Restaurant_Count',
    y='Avg_Rating_Cuisine',
    size='Avg_Price_Cuisine',
    hue='Total_Votes',
    palette='coolwarm',
    alpha=0.8
)
plt.title('Cuisine Market Saturation')
plt.xlabel('Number of Restaurants (Competition)')
plt.ylabel('Average Rating (Quality)')
plt.show()
```

1)Most cuisines have low competition

The majority of cuisines are clustered near the left side (0-20 restaurants).

This means most cuisine types are served by few restaurants â†’ low saturation.

Opportunity: New or niche cuisines still have open market space for entry.

2)High demand (Total Votes) concentrates in few cuisines

Only a few bubbles show warm colors (high Total Votes),
meaning customer demand is concentrated in select cuisines (like North Indian, Fast Food).

 Opportunity: Identify similar cuisines with fewer competitors to capture unmet demand.

```python
cuisine_sat.rename(columns={'Restaurant_Name': 'Restaurant_Count'}, inplace=True)

cuisine_top = cuisine_sat.sort_values(
    by=['Avg_Rating_Cuisine', 'Total_Votes', 'Restaurant_Count'],
    ascending=[False, False, True]
).head(10)

cuisine_top
```

##Price Point Analysis


```python
top_cuisines = df.groupby('Cuisine').Total_Votes.sum().nlargest(5).index
sns.boxplot(data=df[df['Cuisine'].isin(top_cuisines)],
            x='Cuisine', y='Price_per_Vote')
plt.title('Price per Vote Distribution - Top 5 Cuisines')
plt.show()
```


All top cuisines share a similar median price-per-vote (â‚¹50â€“â‚¹150), representing mid-range affordability.

Desserts and Beverages show the widest price range, meaning customers are open to both budget and premium pricing.

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=df[df['Cuisine'].isin(top_cuisines)],
            x='Cuisine', y='Log_Price', palette='coolwarm')
plt.title('Log Price Distribution - Top 5 Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Log Price')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

The Log Price distribution shows a clear mid-market equilibrium, with Desserts, Beverages, and Sichuan offering broad pricing flexibility and potential for premium expansion.

Meanwhile, Fast Food and Pizza remain mass-market, low-margin categories best suited for high-volume delivery models.

##Risk Assesment

```python
risk_df = df.groupby('City')[['Is_Highly_Rated', 'Is_Expensive']].sum().reset_index()
risk_df['Total_Restaurants'] = df.groupby('City')['Restaurant_Name'].nunique().values

# Calculate proportions for easier comparison
risk_df['High_Rated_%'] = (risk_df['Is_Highly_Rated'] / risk_df['Total_Restaurants']) * 100
risk_df['Expensive_%'] = (risk_df['Is_Expensive'] / risk_df['Total_Restaurants']) * 100

risk_df = risk_df.sort_values(by='High_Rated_%', ascending=False)
risk_df
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='City', y='High_Rated_%', data=risk_df, color='seagreen', label='% Highly Rated')
sns.barplot(x='City', y='Expensive_%', data=risk_df, color='salmon', alpha=0.7, label='% Expensive')

plt.title('Risk Assessment: % of Highly Rated vs Expensive Restaurants (Top 10 Cities)')
plt.ylabel('Percentage (%)')
plt.xlabel('City')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

High-High (High Rated + High Expensive): Luxury, competitive â€” requires brand differentiation (e.g., Delhi, Chennai).

High Rated + Low Expensive: Perfect for affordable quality ventures (e.g., Raipur, Lucknow).

Low Rated + High Expensive: Risky but high potential for disruption (e.g., Ulsoor, Goa).

Balanced Markets: Focus on operational efficiency and steady growth (e.g., Pune, Bangalore).

##Channel Disparity

```python
df['Dining_Rating'].median()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df[['Dining_Rating','Delivery_Rating']], palette='Set2')
plt.title('Channel Disparity: Dining vs Delivery Ratings')
plt.ylabel('Rating')
plt.xlabel('Channel')
plt.show()
```

Median Rating (Dining â‰ˆ 3.8) is slightly lower than Delivery (â‰ˆ 4.0).

Delivery ratings are more consistent, with fewer extreme low values compared to dining.

This suggests that customers are generally more satisfied with delivery experiences, possibly due to convenience and faster service.

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df[['Dining_Votes','Delivery_Votes']], palette='mako')
plt.title('Channel Disparity: Dining vs Delivery Votes')
plt.ylabel('Number of Votes')
plt.xlabel('Channel')
plt.show()
```

Dining Votes show a much wider spread and many outliers, indicating some restaurants get huge in-person engagement.

Delivery Votes are mostly clustered near the lower range â€” suggesting fewer customers bother rating delivery, even though the ratings are higher.

Interpretation: Dining attracts more active reviewers, but delivery dominates in satisfaction consistency.

```python
channel_summary = pd.DataFrame({
    'Metric': ['Average Rating', 'Average Votes'],
    'Dining': [df['Dining_Rating'].mean(), df['Dining_Votes'].mean()],
    'Delivery': [df['Delivery_Rating'].mean(), df['Delivery_Votes'].mean()]
})
display(channel_summary)
```

##Optimal Channel Focus

The delivery channel shows more consistent satisfaction levels (less variance, higher average ratings).

The dining channel has higher customer engagement but also higher variability â€” meaning greater risk in maintaining uniform quality.

Customers seem more satisfied with delivery experiences, possibly due to faster service, convenience, and standardization.

Begin as a cloud kitchen focusing on strong packaging, timely delivery, and food quality.

Once a loyal delivery base and strong brand reputation are established, expand into dine-in at key high-demand localities.
